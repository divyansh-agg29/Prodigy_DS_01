<h2>Importing Libraries and the Data</h2>

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
df_1 = pd.read_csv("total_population.csv")
df_2 = pd.read_csv("female_percentage.csv")
df_3 = pd.read_csv("working_age_population_percentage.csv")
```

<h2>Data Cleaning</h2>

First we remove all rows except the ones with country as the 'World'

```python
df_1 = df_1[ df_1["Country Name"].isin(["World"]) ] 
df_2 = df_2[ df_2["Country Name"].isin(["World"]) ]
df_3 = df_3[ df_3["Country Name"].isin(["World"]) ]
```

We remove extranious columns

```python
df_1.drop(columns=['Country Name','Country Code','Indicator Name','Indicator Code','Unnamed: 67'], inplace=True)
df_2.drop(columns=['Country Name','Country Code','Indicator Name','Indicator Code','Unnamed: 67'], inplace=True)
df_3.drop(columns=['Country Name','Country Code','Indicator Name','Indicator Code','Unnamed: 67'], inplace=True)
```

Sample data after cleaning steps: 

```python
df_1.head()
```

```python
df_1.columns
```

Now we will merge the three seperate dataframes into one to ease our later processes

```python
df_1 = pd.melt(df_1, var_name='Year', value_name='Population')
df_1['Year'] = df_1['Year'].astype(int)

df_2 = pd.melt(df_2, var_name='Year', value_name='Female Ratio')
df_2['Year'] = df_2['Year'].astype(int)

df_3 = pd.melt(df_3, var_name='Year', value_name='Working Age Ratio')
df_3['Year'] = df_3['Year'].astype(int)
```

```python
df = df_1.copy()
df['Female Ratio'] = df_2['Female Ratio']
df['Working Age Ratio'] = df_3['Working Age Ratio']
df
```

<h2>Data Visualization</h2>

```python
plt.figure(figsize=(6,4))
plt.plot(df['Year'],df['Population']/(10**9))
plt.xlabel('Year',fontsize=12)
plt.ylabel('World Population (in Billions)')
plt.show()
```

```python
plt.style.use('dark_background')
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
plt.ylim(55,70)
plt.xlabel('Year',fontsize=12)
plt.ylabel('Percentage of Working Class People')
plt.plot(df['Year'],df['Working Age Ratio'],color="blue",linewidth=2)


plt.subplot(1,2,2)
plt.ylim(49.6,50.2)
plt.xlabel('Year',fontsize=12)
plt.ylabel('Percentage of Females')
plt.plot(df['Year'],df['Female Ratio'],color="red",linewidth=2)
plt.plot(df['Year'],[50]*63,color="white",linestyle="dotted",linewidth=1)
plt.show()
```
